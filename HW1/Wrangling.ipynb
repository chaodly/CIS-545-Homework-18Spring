## Step 4.2.1

```python
# TODO: output airports you can fly to by not from

#left = pd.DataFrame(airports_df['airport_id'])
# must transpose, because pd.DataFrame([,,,]) is a line

entire = pd.DataFrame([airports_df['airport_id'], airports_df['name'], airports_df['city'], airports_df['country'], airports_df['iata/faa']]).transpose()
destination = pd.DataFrame(routes_df['destination_airport_id'])
source = pd.DataFrame(routes_df['source_airport_id'])

# those airports that are both destination and source are included
temp = pd.merge(entire, destination, left_on = 'airport_id', right_on = 'destination_airport_id', how = 'inner').drop_duplicates()

destination_airports = pd.merge(temp, source, left_on = 'destination_airport_id', right_on = 'source_airport_id', how = 'outer').drop_duplicates()

destination_only = destination_airports[destination_airports['source_airport_id'].isnull()].drop('source_airport_id', axis = 1)

# reset the indexes at the original place (inplace = True), and discard the original indexes(drop = True).
destination_only.reset_index(drop = True, inplace = True)

# destination_only
```

## Step 4.2.2

```python
#TODO: top 10 most popular airports
entire = pd.DataFrame([airports_df['airport_id'], airports_df['name'], airports_df['city'], airports_df['country']]).transpose()
destination = pd.DataFrame(routes_df['destination_airport_id'])

new = pd.merge(destination, entire, left_on = 'destination_airport_id', right_on = 'airport_id', how = 'inner',)

ranking = new.drop_duplicates().set_index('airport_id').drop(labels = 'destination_airport_id', axis = 1)
count = new['airport_id'].value_counts()

count = pd.DataFrame(count).rename(columns = {'airport_id' : 'counts'})
count.index.name = 'airport_id'
ranking['counts'] = count

ranking.sort_values(by = 'counts', ascending = False, inplace = True)
ranking = ranking.head(10).set_index([list(range(1, 11))]) # range(1,11) = 1 : 10
# ranking
```

## Step 4.3

```python
# TODO: Create dataframe of airline_iata, number of incidents, number of routes
route_count = pd.DataFrame(routes_df['airline'])

iata = pd.DataFrame(airlines_df['iata'])
airline_counts = pd.DataFrame(route_count['airline'].value_counts()).reset_index().rename(columns = {'index' : 'airline', 'airline':'counts'})

iata_counts_df = pd.merge(iata, airline_counts, left_on = 'iata', right_on = 'airline', how = 'inner').drop('iata', axis = 1)

match_by_name_df = pd.merge(airlines_df, incidents_df, left_on = 'name', right_on = 'Airline', how = 'inner')
match_by_alias_df = pd.merge(airlines_df, incidents_df, left_on = 'alias', right_on = 'Airline', how = 'inner')

total_incidents_df = match_by_name_df.append(match_by_alias_df, ignore_index = True)
incident_counts_df = pd.DataFrame(total_incidents_df['iata'].value_counts().reset_index().rename(columns = {'index':'incident_iata', 'iata': 'counts'}))

incident_route_iata_df = pd.merge(incident_counts_df, iata_counts_df, left_on = 'incident_iata', right_on = 'airline', how = 'inner').drop('airline', axis = 1)
incident_route_iata_df.rename(columns = {'incident_iata':'airline_iata', 'counts_x' : 'incident_counts', 'counts_y':'route_counts'}, inplace = True)
incident_route_iata_df
```

```python
# TODO: Scatter plot of number of routes vs. number of incidents
import matplotlib.pyplot as plt
result = incident_route_iata_df[incident_route_iata_df['incident_counts'] > 2]
plot = plt.scatter(x = result['route_counts'], y = result['incident_counts'])
plt.xlabel('Route Counts')
plt.ylabel('Incident Counts')
plt.title('numbers of routes vs. numbers of incidents'.title())

for i in range(len(result)):
    plot.axes.annotate(result.ix[i, 0], [result.ix[i, 2],result.ix[i, 1]])
```

```python
# TODO: get random sample of flights from SQL
flights_s_df = pd.read_sql_query('SELECT * FROM flights_df ORDER BY RANDOM() LIMIT 1000', con = engine)
```

```python
# TODO: create nested map of Jaccard distance
nested_map = dict()
for col in routes_df:
    if routes_df[col].dtype == object:
        nested_map[col] = dict()
        for row in flights_s_df:
            if flights_s_df[row].dtype == object:
                flight_set = set(flights_s_df[row])
                routes_set = set(routes_df[col])
                u = flight_set.union(routes_set)
                i = flight_set.intersection(routes_set)
                if len(i)/len(u):
                    nested_map[col][row] = len(i)/len(u)
# nested_map
```
