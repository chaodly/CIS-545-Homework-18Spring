```python
# fillna
# Replace NaNs with blanks if the column is a string, so we donâ€™t get
# objects in the column
def fillna_col(series):
    if series.dtypes is pd.np.dtype(object):
        return series.fillna('')
    else:
        return series
```

```python
# TODO: Define nullify here
# Replace \Ns with NaNs
def nullify(x):
    if x == r"\N":
        return float('nan')
    else:
        return x
```

```python
# to NaN
airlines_df = airlines_df.applymap(nullify)
airports_df = airports_df.applymap(nullify)
routes_df = routes_df.applymap(nullify)

# NaN to blank
airlines_df = airlines_df.apply(fillna_col)
airports_df = airports_df.apply(fillna_col)


# drop blank rows
routes_df = routes_df.dropna(subset = ['airline_id', 'source_airport_id', 'destination_airport_id'])
routes_df = routes_df.apply(fillna_col)
# print (routes_df)
```

```python
routes_df['airline_id'] = routes_df['airline_id'].astype(int)
routes_df['source_airport_id'] = routes_df['source_airport_id'].astype(int)
routes_df['destination_airport_id'] = routes_df['destination_airport_id'].astype(int)
```

```python
# TODO: read and clean raw text file and store in clean_incidents
clean_incidents = []
lines = []
with open ("incidents_raw.txt", "r") as stream:
    while True:
        line = stream.readline()
        if line:
            lines.append(line.strip("[edit]\n"))  
        else:
            break
    
lines = lines[lines.index('1997') : ]

incident_1997 = []

for item in lines:
    incident_1997.append(item.replace("â€“", ","))

year = 1997
incident_1997.append('2019')
while year <= 2018:
    start = str(year)
    start_index = incident_1997.index(start) + 1
    end_index = incident_1997.index(str(year + 1))
    for i in range(start_index, end_index):
        if ('Flight' in incident_1997[i]):
            clean_incidents.append(start + ' ' + incident_1997[i])
       
    year += 1
    
# clean_incidents
```

```python
# TODO: convert clean_incidents into dataframe incidents_df
incidents_df = pd.DataFrame(columns = ['Date', 'Airline', 'FlightNum'])

for incident in clean_incidents:
    date = incident[: incident.find(',') - 1]
    
    index = incident.find('Flight')
    airline_name = incident[incident.rfind(',', 0, index) + 2 : index - 1]
    start = index + 7
    end = start
    while incident[end].isdecimal():
        end += 1
        
    flight_number = incident[start : end]
    
    entity = {'Date' : date, 'Airline' : airline_name, 'FlightNum' : flight_number}
    incidents_df = incidents_df.append(entity, ignore_index = True)

# incidents_df
```

```python
# TODO: change column type of 'Date'
incidents_df['Date'] = pd.to_datetime(incidents_df['Date'])
```
